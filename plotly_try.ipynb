```python
import plotly.express as px
import pandas as pd
import numpy as np

# Sample data for a 3D helix
t = np.linspace(0, 10 * np.pi, 100)
x = np.sin(t)
y = np.cos(t)
z = t

df = pd.DataFrame({"x": x, "y": y, "z": z})

# Create the 3D line plot
fig = px.line_3d(df, x="x", y="y", z="z", title="3D Line Plot with Plotly Express")

# Display the figure
fig.show()
```

```python
import plotly.express as px

df = px.data.gapminder().query("year == 2007")

fig = px.scatter_3d(
    df, x="gdpPercap", y="lifeExp", z="pop",
    color="continent", size="pop",
    hover_name="country", size_max=40,
    log_x=True, log_z=True,
    template="plotly_dark",
    title="🌍 GDP vs Life Expectancy vs Population (2007)"
)

fig.update_traces(marker=dict(opacity=0.8, line=dict(width=0)))
fig.update_layout(scene=dict(
    xaxis_title="GDP per Capita (log)",
    yaxis_title="Life Expectancy",
    zaxis_title="Population (log)"
))

fig.show()
```
