```python
# import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats
```

```python
iris = pd.read_csv('/home/codespace/.cache/kagglehub/datasets/uciml/iris/versions/2/Iris.csv')

iris.head()
```

```python
iris.describe()
```

```python
sns.set(style="whitegrid")
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
```

## Bar Chart

```python
iris[features].mean().plot(kind='bar', color='skyblue')
plt.title('Average Feature Values')
plt.ylabel('Mean Value')
plt.show()
```

## Stacked Bar Chart

```python
iris.groupby('Species')[features].mean().plot(
    kind='bar',
    stacked=True,
    figsize=(8,5)
)
plt.title('Stacked Bar Chart of Feature Means by Species')
plt.ylabel('Mean Value')
plt.show()
```

## Line Graph

```python
for feature in features:
    plt.plot(iris[feature], label=feature)

plt.legend()
plt.title('Line Graph of Features')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```

## Histograpm

```python
iris['SepalLengthCm'].hist(bins=20, color='orange')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

## Scatter plot

```python
sns.scatterplot(
    data=iris,
    x='SepalLengthCm',
    y='PetalLengthCm',
    hue='Species'
)
plt.title('Scatter Plot: Sepal Length vs Petal Length')
plt.show()
```

## Box plot

```python
sns.boxplot(
    data=iris,
    x='Species',
    y='PetalLengthCm'
)
plt.title('Box Plot of Petal Length by Species')
plt.show()
```

## Parallel Coordinate Plot

```python
from pandas.plotting import parallel_coordinates
```

```python
parallel_coordinates(
    iris[features + ['Species']],
    'Species',
    colormap='Set1'
)
plt.title('Parallel Coordinate Plot')
plt.show()
```

##  Heatmap

```python
corr = iris[features].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)
plt.title('Correlation Heatmap')
plt.show()
```

## QQ Plot

```python
stats.probplot(iris['SepalLengthCm'], plot=plt)
plt.title('QQ Plot of Sepal Length')
plt.show()
```

```python
sns.violinplot(
    data=iris,
    x='Species',
    y='PetalWidthCm'
)
plt.title('Violin Plot of Petal Width by Species')
plt.show()
```

## Pair Plot

```python
sns.pairplot(
    iris,
    hue='Species'
)
plt.show()
```

## KDE Plot

```python
sns.kdeplot(
    data=iris,
    x='SepalLengthCm',
    hue='Species',
    fill=True
)
plt.title('KDE Plot of Sepal Length')
plt.show()
```

## Bubble Chart

```python
plt.scatter(
    iris['SepalLengthCm'],
    iris['SepalWidthCm'],
    s=iris['PetalWidthCm'] * 100,
    c=pd.factorize(iris['Species'])[0],
    alpha=0.6
)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Bubble Chart')
plt.show()
```
