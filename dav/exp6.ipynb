## 1. Import Reqd Libraries

```python
import pandas as pd
import geopandas as gpd
import folium
from folium.plugins import HeatMap
```

## 2. Load dataset

```python
df = pd.read_csv('data/india_20_cities_dataset.csv')
gdf = gpd.read_file('data/india_states_simplified.geojson')

print(f"DataFrame top 5: \n{df.head()}")
print(f"\nDataFrame shape: {df.shape}")
print(f"\n\nGeoDataFrame top 5: \n{gdf.head()}")
print(f"\nGeoDataFrame shape: {gdf.shape}")
```

## 3. Create base map

```python
m = folium.Map(
    location=[20.5937, 78.9629],
    zoom_start=5
)

folium.GeoJson(gdf).add_to(m)
```

## 4. Create marker map

```python
for i, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['City'],
        tooltip=row['City']
    ).add_to(m)
```

## 5. Create Choropleth Map

```python
folium.Choropleth(
    geo_data=gdf,
    name='choropleth',
    data=df,
    columns=['State', 'Cases'],   # example columns
    key_on='feature.properties.state',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Cases by State'
).add_to(m)
```

## 6. Add Multiple Dataset Layers

```python
heat_data = [[row['Latitude'], row['Longitude']] for index, row in df.iterrows()]

HeatMap(heat_data).add_to(m)
```

## 7. Display Map

```python
m.save("output_map.html")

m
```
